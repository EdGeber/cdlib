```python
%load_ext autoreload
%autoreload 2
import nclib
import networkx as nx
from nclib import community, ensemble, viz
import seaborn as sns
import numpy as np
import matplotlib as plt
import pandas as pd
%matplotlib inline
```

```python
graphs = []
for i in np.arange(2,50,10):
    for j in range(2):
        g = nx.planted_partition_graph(i,10,0.5,0.1)
        g.graph["name"]=i
        graphs.append(g)
```

```python
resolution = ensemble.Parameter(name="resolution", start=0.5, end=1, step=0.1)

part_ref = []
part = []
for g in graphs:
    reference = nclib.NodeClustering(communities=g.graph["partition"],graph=g,method_name="reference")
    
    pooled = ensemble.pool(g,[community.louvain],configurations=[[resolution]])

    for communities in pooled:
        part_ref.append(reference)
        part.append(communities)
```

```python
def plot_scoring(scoring,partitions_ref,partitions_to_eval):
    forDF = []
    for i,p in enumerate(partitions_to_eval):
        g = p.graph
        solutionDesciption= str(p.method_name)+"|"+str(p.method_parameters)
        score = scoring(partitions_ref[i],p)
        forDF.append([g.graph["name"],score,solutionDesciption])
    df = pd.DataFrame(columns=["graph name","score","methodName"],data=forDF)
    ax = sns.lineplot(x="graph name",y="score",hue="methodName",data=df,legend="brief")
    ax.legend(loc='center left', bbox_to_anchor=(1.25, 0.5), ncol=1)
            
```

```python
g = nx.planted_partition_graph(10,10,0.5,0.1)
reference = nclib.NodeClustering(communities=g.graph["partition"],graph=g,method_name="reference")
coms = community.louvain(g=g)
coms2 = community.infomap(g=g)
coms3 = community.eigenvector(g=g)
```

```python
def plot_com_stat(com_clusters,com_property):
    allVals = []
    allNames = []
    for c in com_clusters:
        prop = com_property(c.graph,c,summary=False)
        allVals+= prop
        allNames+= [c.method_name+str(c.method_parameters)]*len(prop)

    ax = sns.violinplot(allNames,allVals)
    for tick in ax.get_xticklabels():
        tick.set_rotation(90)
```

```python
plot_com_stat([coms2,coms3,reference],nclib.evaluation.internal_edge_density)
```
