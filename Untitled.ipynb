```python
def plot_com(com_clusters,com_fitness):
    """
    Plot a distribution of a property among communities for a set of clusterings. (violin-plot)

    :param com_clusters: set of clusterings to compare
    :param com_fitness: the fitness/community property to use
    """
    allVals = []
    allNames = []
    for c in com_clusters:
        prop = com_fitness(c.graph,c,summary=False)
        allVals+= prop
        allNames+= [c.get_description()]*len(prop)

    ax = sns.violinplot(allNames,allVals)
    for tick in ax.get_xticklabels():
        tick.set_rotation(90)

    return ax
```
